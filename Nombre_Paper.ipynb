````
#https://scikit-learn.org/stable/modules/preprocessing.html
#https://scikit-learn.org/stable/auto_examples/preprocessing/plot_all_scaling.html

#buscamos visualmente
def grafica5(datos, titulo):
  fig = plt.figure(constrained_layout=True, figsize=[15, 10])

  gs = GridSpec(2, 3, figure=fig)
  ax1 = fig.add_subplot(gs[0, 0])
  # identical to ax1 = plt.subplot(gs.new_subplotspec((0, 0), colspan=3))
  ax2 = fig.add_subplot(gs[0, 1])
  ax3 = fig.add_subplot(gs[0, -1])
  ax4 = fig.add_subplot(gs[-1, 0])
  ax5 = fig.add_subplot(gs[-1, -1])

  raro = sns.color_palette(['blue', 'red', 'cyan', 'gray', 'pink', 'lime'])

  sns.scatterplot(x='CLASS', y='SHIP2', data=datos, ax=ax1, hue='CLASS', style='CLASS', palette=raro, s=120, alpha=0.7, edgecolors='none', legend=False)
  sns.scatterplot(x='CLASS', y='PTEN', data=datos, ax=ax2, hue='CLASS', style='CLASS', palette=raro, s=120, alpha=0.7, edgecolors='none', legend=False)
  sns.scatterplot(x='CLASS', y='XIAP', data=datos, ax=ax3, hue='CLASS', style='CLASS', palette=raro, s=120, alpha=0.7, edgecolors='none', legend=False)
  sns.scatterplot(x='CLASS', y='PI3K', data=datos, ax=ax4, hue='CLASS', style='CLASS', palette=raro, s=120, alpha=0.7, edgecolors='none', legend=False)
  sns.scatterplot(x='CLASS', y='AKT', data=datos, ax=ax5, hue='CLASS', style='CLASS', palette=raro, s=120, alpha=0.7, edgecolors='none')

  # Tweak the visual presentation  
  #ax.xaxis.grid(True)  
  #ax.set(ylabel="")  
  #sns.despine(trim=True, left=True)
  fig.suptitle(titulo)
  h,l=ax5.get_legend_handles_labels() # get labels and handles from ax1
  fig.legend(h,l,loc='center', bbox_to_anchor=(.5, .25))
  ax5.get_legend().remove()
  fig.show()
  #g.set(ylim=(-1, 11), yticks=[0, 5, 10]);

def escalar(datos, tecnica=1):
  reproducibilidad()
  x_scaled = None
  if tecnica ==1:
    #('Data after MinMax scaling',
    x_scaled = MinMaxScaler().fit_transform(datos)
  elif tecnica == 2:
    #('Data after robust scaling',
    x_scaled = RobustScaler(quantile_range=(25, 75)).fit_transform(datos)
  elif tecnica == 3:
    #('Data after power transformation (Yeo-Johnson)',
    x_scaled = PowerTransformer(method='yeo-johnson').fit_transform(datos)
  elif tecnica == 4:
    #('Data after quantile transformation (gaussian pdf)',
    x_scaled = QuantileTransformer(output_distribution='normal').fit_transform(datos)
  elif tecnica == 5:
    #('Data after quantile transformation (uniform pdf)',
    x_scaled = QuantileTransformer(output_distribution='uniform').fit_transform(datos)
  elif tecnica == 6:
    #('Data after sample-wise L2 normalizing',
    x_scaled = Normalizer().fit_transform(datos)
  return x_scaled

grafica5(df, 'Outlier inspection')
````

````
grafica5(n_df, 'Power Transformation (Yeo-Johnson)')
````

##Eliminar outliers mediante Z-score

````
threshold = 3
df_z = df.drop('CLASS', axis=1).values
z = np.abs(stats.zscore(df_z))
df_z = df_z[(z < threshold).all(axis=1)]
df_z =pd.DataFrame(df_z, columns=df.columns[:-1])
df_z['CLASS']=df['CLASS']

n_df_z = n_df.drop('CLASS', axis=1).values
z = np.abs(stats.zscore(n_df_z))
n_df_z = n_df_z[(z < threshold).all(axis=1)]
n_df_z =pd.DataFrame(n_df_z, columns=df.columns[:-1])
n_df_z['CLASS']=df['CLASS']

grafica5(df, 'original')
grafica5(df_z, 'Z-score outlier removal')
grafica5(n_df_z, 'scaled Z-score outlier removal')
````

##Eliminar outliers mediante i-Forest

````
#https://machinelearningmastery.com/model-based-outlier-detection-and-removal-in-python/
#https://scikit-learn.org/stable/modules/outlier_detection.html
#https://scikit-learn.org/stable/auto_examples/miscellaneous/plot_anomaly_comparison.html
iso = IsolationForest(contamination=0.1)
yhat = iso.fit_predict(df.drop('CLASS', axis=1))
mask = yhat != -1
df_i=pd.DataFrame(df.values[mask,:], columns=df.columns)

yhat = iso.fit_predict(n_df.drop('CLASS', axis=1))
mask = yhat != -1
n_df_i=pd.DataFrame(n_df.values[mask,:], columns=n_df.columns)

grafica5(df, 'Original')
grafica5(df_i, 'i-Forest outlier removal')
grafica5(n_df_i, 'scaled i-Forest outlier removal')
````

##Eliminar outliers mediante Local Outlier Factor

````
#https://machinelearningmastery.com/model-based-outlier-detection-and-removal-in-python/
#https://scikit-learn.org/stable/modules/outlier_detection.html
#https://scikit-learn.org/stable/auto_examples/miscellaneous/plot_anomaly_comparison.html

lof = LocalOutlierFactor(n_neighbors=31, contamination=.1)
yhat = lof.fit_predict(df.drop('CLASS', axis=1))
mask = yhat != -1
df_l=pd.DataFrame(df.values[mask,:], columns=df.columns)

yhat = lof.fit_predict(n_df.drop('CLASS', axis=1))
mask = yhat != -1
n_df_l=pd.DataFrame(n_df.values[mask,:], columns=n_df.columns)

grafica5(df, 'Original')
grafica5(df_l, 'Local Outlier Factor')
grafica5(n_df_l, 'scaled Local Outlier Factor')
````

##Utilizando Cross validation

**Problema, solo probamos con un conjunto de validación**

**Solución, validación cruzada (cross validation)** (estratificada)
This cross-validation object is a variation of KFold that returns stratified folds. The folds are made by preserving the percentage of samples for each class.

```
         Split training samples into N folds

 +-------+   +-------+   +-------+         +-------+
 |       |   |       |   |       |         |       |
 |   1   |   |   2   |   |   3   |   ...   |   N   |
 |       |   |       |   |       |         |       |
 +-------+   +-------+   +-------+         +-------+

Take one fold as validation set and train on N-1 folds

 +-------+   +-------+   +-------+         +-------+
 |*******|   |       |   |       |         |       |
 |*******|   |   2   |   |   3   |   ...   |   N   |
 |*******|   |       |   |       |         |       |
 +-------+   +-------+   +-------+         +-------+

         Take the next one as validation set

 +-------+   +-------+   +-------+         +-------+
 |       |   |*******|   |       |         |       |
 |   1   |   |*******|   |   3   |   ...   |   N   |
 |       |   |*******|   |       |         |       |
 +-------+   +-------+   +-------+         +-------+

          Repeat the procedure for all folds

 +-------+   +-------+   +-------+         +-------+
 |       |   |       |   |       |         |*******|
 |   1   |   |   2   |   |   3   |   ...   |*******|
 |       |   |       |   |       |         |*******|
 +-------+   +-------+   +-------+         +-------+

            And average out the accuracy
```
Referencia dibujo https://colab.research.google.com/drive/1My8UggN12Opt_gscK3tl4VLhZkHiQSyX#scrollTo=AM5yUx2luMw-

````
#The mean score and the 95% confidence interval of the score estimate are returned

def valCruzada(X, Y, nof_folds=10, max_k=85):
  reproducibilidad()

  mean_scores = []
  std_conf = []

  #nof_folds = 10 #10-fold cross validation

  for k in range(1, max_k):
    clf = KNeighborsClassifier(n_neighbors=k)

    # cross-validate knn on our training sample with nof_folds
    cv_generator = StratifiedKFold(10, True)
    #Regresa a dict of arrays containing test_score, train_score, fit_time, score_time, estimator
    scores = cross_validate(clf, X, Y, scoring=['accuracy'], cv=cv_generator)
    #solo interesa el test_accuracy
    scores = scores['test_accuracy']
    # add avg accuracy to score table
    mean_scores.append(scores.mean())
    #95% confianza = margin of error https://en.wikipedia.org/wiki/68%E2%80%9395%E2%80%9399.7_rule
    std_conf.append(scores.std() * 2)

  return mean_scores, std_conf
````
